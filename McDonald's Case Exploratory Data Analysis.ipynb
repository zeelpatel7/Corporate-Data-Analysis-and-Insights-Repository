### 2.4 Understanding the Data Quality

#### 2.4.1 Outliers in Weekly Sales Data

```python
print("\nOutliers in Weekly Sales Data:")
for column in weekly_sales.columns:
    if weekly_sales[column].dtype != 'object':
        Q1 = weekly_sales[column].quantile(0.25)
        Q3 = weekly_sales[column].quantile(0.75)
        IQR = Q3 - Q1
        outliers = weekly_sales[(weekly_sales[column] < (Q1 - 1.5 * IQR)) | (weekly_sales[column] > (Q3 + 1.5 * IQR))]
        print(f"Outliers in {column}:")
        print(outliers)
        # Plotting the outliers
        plt.figure(figsize=(10, 6))
        sns.boxplot(data=weekly_sales, x=column, color='red')
        label = column_labels.get(column, column)
        plt.title(f'Outliers in {label} (Weekly Sales Data)')
        plt.xlabel(label)
        plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=weekly_sales.select_dtypes(include=['float64', 'int64']), color = 'red' )
plt.title('Boxplot for Outlier Detection in Weekly Sales Data')
plt.show()
```

### 2.6 Identify Clusters or Subgroups

#### 2.6.1 K-means Cluster

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Select relevant features for clustering
features_for_clustering = weekly_sales[['latitude', 'longitude', 'wavg_price', 'urws', 'upt', 'agc']].dropna()

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(features_for_clustering)

# Perform K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)  # Choose the appropriate number of clusters
kmeans.fit(scaled_data)
features_for_clustering['Cluster'] = kmeans.labels_

# Visualize clusters (using latitude and longitude for simplicity)
plt.figure(figsize=(10, 6))
plt.scatter(features_for_clustering['latitude'], features_for_clustering['longitude'], c=features_for_clustering['Cluster'], cmap='viridis')
plt.title('Clusters of Weekly Sales Data')
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.show()
```

## 3. Hypothesis Testing

### 3.1 Hypothesis Test 1: Impact of Promotion on Total Sales

```python
from scipy import stats

# Replace 'YYYY-MM-DD' with the actual start date of the promotion
promotion_start_date = '2015-10-06'  # Example date

# Split the data into pre-promotion and post-promotion
pre_promotion = weekly_sales[weekly_sales['wk_ending'] < promotion_start_date]['totunits'].dropna()
post_promotion = weekly_sales[weekly_sales['wk_ending'] >= promotion_start_date]['totunits'].dropna()

# Ensure both groups have data
if pre_promotion.empty or post_promotion.empty:
    print("One of the groups has no data. Please check the promotion start date or the dataset.")
else:
    # Perform a t-test
    t_stat, p_value = stats.ttest_ind(post_promotion, pre_promotion)
    print(f"T-statistic: {t_stat}, P-value: {p_value}")

    # Interpretation
    if p_value < 0.05:
        print("Result: Reject the null hypothesis (significant difference found).")
    else:
        print("Result: Fail to reject the null hypothesis (no significant difference found).")

```
